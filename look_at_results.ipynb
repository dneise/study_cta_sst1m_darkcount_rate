```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_hdf('results.h5')
for path, _ in df.groupby('path'):
    plt.figure(figsize=(12, 4))
    plt.title(path)
    for method, sub in _.groupby('method'):
        plt.plot(
            sub.pixel_id, 
            sub.dark_count_rate_MHz, 
            '.:',
            label=method
        )
    plt.ylim(0, 5)
```
